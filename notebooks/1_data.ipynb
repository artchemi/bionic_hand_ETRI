```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = loadmat('s1/S1_E2_A1.mat')
data.keys()
```

```python
np.unique(data['stimulus'], return_counts=True)
```

```python
data['sensor']
```

```python
data['emg']
```

```python
# Настройка стиля seaborn
sns.set(style="whitegrid")
# Цвета — пастельные
colors = sns.color_palette("pastel")

fig, axs = plt.subplots(2, 1, figsize=(8, 6), sharex=True)  # 2 строки, 1 столбец

axs[0].plot(data['stimulus'][:10000])
axs[0].set_title('Gesture timeline')

axs[1].plot(data['emg'][:10000][:,0])
axs[1].set_title('sEMG timeline')
axs[1].set_xlabel('Time, ms')

plt.tight_layout()
plt.show()
```
